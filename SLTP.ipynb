```python
SLl=[0] * len(pos)
TPl=[0] * len(pos)
SL = 0.0050
TP = 0.0100
for i in range (len(pos)):
    if pos.Operaciones[i-1] == 'sell':
        SLl[i] = round(pos.open[i] + SL,4)
        TPl[i] = round(pos.open[i] - TP,4)
    elif pos.Operaciones[i-1] == 'buy':
        SLl[i] = round(pos.open[i] - SL,4)
        TPl[i] = round(pos.open[i] + TP,4)
pos['SL'] = SLl
pos['TP'] = TPl
pos
```

```python
intra['Fecha'] = intra.index.date
```

```python
pos['result'] = [0] * len(pos)
```

```python
for n in range (len(pos)):
    for i in range (len(intra)):
        if intra.Fecha[i]== pos.index[n]:
            if SLl[n] == intra.open[i]:
                pos['result'][n] = 1
            elif SLl[n] == intra.close[i]:
                pos['result'][n] = 1
            elif SLl[n] == intra.high[i]:
                pos['result'][n] = 1
            elif SLl[n] == intra.low[i]:
                pos['result'][n] = 1
            elif TPl[n] == intra.open[i]:
                pos['result'][n] = 2
            elif TPl[n] == intra.close[i]:
                pos['result'][n] = 2
            elif TPl[n] == intra.high[i]:
                pos['result'][n] = 2
            elif TPl[n] == intra.low[i]:
                pos['result'][n] = 2
```

```python
pos
```

```python
R = [0] * len(pos)
for n in range (len(pos)):
    if pos.Operaciones[n] == 'buy':
        if pos['result'][n] == 0:
            R[n] = pos.close[n] - pos.open[n]
        elif pos['result'][n] == 1:
            R[n] = SLl[n] - pos.open[n]
        elif pos['result'][n] == 2:
            R[n] = TPl[n] - pos.open[n]
    elif pos.Operaciones[n] == 'sell':
        if pos['result'][n] == 0:
            R[n] = pos.open[n] - pos.close[n]
        elif pos['result'][n] == 1:
            R[n] = pos.open[n] - SLl[n]
        elif pos['result'][n] == 2:
            R[n] = pos.open[n] - TPl[n]
```
